```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def transform(x, y, z):
    a = np.asarray([[np.cos(np.radians(z)), -np.sin(np.radians(z))],
                    [np.sin(np.radians(z)), np.cos(np.radians(z))]])
    b = np.asarray([x, y])
    c = np.dot(a, b)
    return c[0], c[1]
```

```python
def plot(x1, y1, x2, y2):
    plt.figure(figsize=(10,10))
    plt.gca().set_aspect('equal')
    plt.xlim([-10, 10])
    plt.ylim([-10, 10])
    plt.plot([9, -9], [0, 0])
    plt.plot([0, 0], [9, -9])
    plt.scatter(x1, y1, s=100)
    plt.scatter(x2, y2, s=100)
    plt.legend(['X', 'Y', 'Original', 'Transformed'])
    plt.show()
```

```python
def spot(x1, y1, x2, y2):
    print('Original: (%f, %f)' % (x1, y1))
    print('Transformed: (%f, %f)' % (x2, y2))
```

```python
def main(x, y, z):
    x2, y2 = transform(x, y, z)
    spot(x, y, x2, y2)
    plot(x, y, x2, y2)
```

```python
main(1, 0, -90)
```
