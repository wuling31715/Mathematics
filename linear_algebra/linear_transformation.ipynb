# 線性轉換

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import random as r
```

```python
an = np.linspace(0, 2 * np.pi, 100)
```

```python
plt.plot(2 * np.cos(an), 2 * np.sin(an))
plt.axis('equal')
```

```python
plt.plot(10 * 2 * np.cos(an), 2 * np.sin(an))
plt.axis('equal')
```

```python
plt.plot(2 * np.cos(an), 10 * 2 * np.sin(an))
plt.axis('equal')
```

```python
plt.plot(2 * np.cos(2 * an), 2 * np.sin(an))
plt.axis('equal')
```

```python
plt.plot(2 * np.cos(an), 2 * np.sin(0.5 * an))
plt.axis('equal')
```

```python
plt.plot(2 * np.cos(1000 * an), 2 * np.sin(an))
plt.axis('equal')
```

```python
plt.plot(2 * np.cos(an), 2 * np.sin(0.001 * an))
plt.axis('equal')
```
