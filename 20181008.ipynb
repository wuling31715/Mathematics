```python
def absolute(x):
    if x < 0:
        return -x
    else:
        return x
```

```python
X = [x for x in range(-100, 100)]
Y = [absolute(x) for x in range(-100, 100)]
plt.plot(X, Y)
plt.show()
```

# Zeta Function

```python
def zeta(k):
    sum = 0
    for i in range(1, 10):
        sum += (sp.S(1) / (i ** k))
    return sum
```

```python
zeta(4)
```

```python
X = []
Y = []
T = np.linspace(1, 100, 100)
```

```python
for i in T:
    p = mp.zeta(0.5 + i * 1j)
    X.append(p.real)
    Y.append(p.imag)
plt.plot(X, Y)
plt.show()
```
