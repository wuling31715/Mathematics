# 作業四：「聽見」sin 函數

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pyaudio
```

### 1. 定義 sin 函數

```python
def sin(frequency, second, sampleRate):
    n = int(second * sampleRate)
    interval = 2 * np.pi * frequency / sampleRate
    X = np.arange(n) * interval
    Y = np.sin(X)
    return X, Y
```

### 2. 設定頻率、波長

```python
def play(stream, frequency=440, second=1, sampleRate=44100):
    X, Y = sin(frequency, second, sampleRate)
    stream.write(Y.astype(np.float32).tostring())
    return X, Y
```

### 3. 播放

```python
p = pyaudio.PyAudio()
stream = p.open(format=pyaudio.paFloat32, channels=1, rate=44100, output=True)
X, Y = play(stream, frequency=300, second=3)
stream.close()
p.terminate()
```

```python
plt.figure(figsize=(20, 10))
plt.plot(X[:1000], Y[:1000])
plt.show()
```
