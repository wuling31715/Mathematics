```python
def A(x):
    if x == 1:
        while True:
            print('A is in loop')
    elif x == 0:
        return 'A is halt'
```

```python
def B(x):
    x = int(x)
    if x == 1:
        A(x)    
        return 'B is halt'
    elif x == 0:
        while True:            
            print('B is in loop')
```

```python
B(1)
```
