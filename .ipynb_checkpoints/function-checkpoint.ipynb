# Function

```python
import matplotlib.pyplot as plt
```

$ f(x) = \left\{ \begin{array}{rr} x^2, & x \geq 0 \\ -x^2, & x < 0 \end{array}\right. $

考慮 $ x = 0 $：

$ \lim_{x \to 0^-}\frac{f(x)-f(0)}{x-0} = \lim_{x \to 0^-}\frac{-x^2}{x} = \lim_{x \to 0^-}-x = 0 $

$ \lim_{x \to 0^+}\frac{f(x)-f(0)}{x-0} = \lim_{x \to 0^+}\frac{x^2}{x} = \lim_{x \to 0^+}x = 0$

$ \rightarrow $ 左極限 ＝ 右極限

$ \rightarrow x = 0 $ 時極限存在

$ \rightarrow f(x) $ 處處可微

```python
def f(x):
    if x >= 0:
        return x ** 2
    else:
        return -(x ** 2)
```

```python
X = [x for x in range(-100, 100)]
Y = [f(x) for x in range(-100, 100)]
plt.figure(figsize=(20, 10))
plt.plot(X, Y)
plt.show()
```

$ g(x) = x^2 $

$ \frac{g(x)}{dx} = 2x $

$ \rightarrow g(x) $ 處處可微

```python
Y2 = [x ** 2 for x in range(-100, 100)]
plt.figure(figsize=(20, 10))
plt.plot(X, Y2)
plt.show()
```

### f(x), g(x)

$ \left\{ \begin{array}{rr} f(x) = g(x), & x \geq 0 \\ f(x) \neq g(x), & x < 0 \end{array}\right. $

```python
plt.figure(figsize=(20, 10))
plt.plot(X, Y)
plt.plot(X, Y2)
plt.show()
```
