```python
import matplotlib.pyplot as plt
from numpy import sin, cos
```

```python
X = [x for x in range(-100, 100)]
S = [sin(x) for x in range(-100, 100)]
plt.figure(figsize=(20, 10))
plt.plot(X, S)
plt.show()
```

```python
X = [x for x in range(-100, 100)]
C = [cos(x) for x in range(-100, 100)]
plt.figure(figsize=(20, 10))
plt.plot(X, C)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.plot(X, S)
plt.plot(X, C)
plt.legend(['sin', 'cos'])
plt.show()
```
