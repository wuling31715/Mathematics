```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
def function(x):
    if x >= 0:
        return x ** 2
    else:
        return - x ** 2
```

```python
X = np.linspace(-10, 10 , 100)
Y = [function(x) for x in X]
plt.plot(X, Y)
plt.show()
```

```python
x = np.linspace(-10, 10 , 100)
y = np.piecewise(x, [x >= 0, x < 0], [lambda x: x ** 2, lambda x : - x ** 2])
plt.plot(x, y)
plt.show()
```
