```python
import matplotlib.pyplot as plt
from numpy import sin, cos
```

```python
X = [x for x in range(100)]
Y = [y for y in range(100)]
plt.figure(figsize=(20, 10))
plt.plot(X, Y)
plt.show()
```

```python
X = [cos(x) for x in range(100)]
Y = [sin(y) for y in range(100)]
plt.figure(figsize=(10, 10))
plt.plot(X, Y)
plt.show()
```

```python
def f(x):
    if x >= 0:
        return x ** 2
    else:
        return -(x ** 2)
```

```python
X = [x for x in range(-100, 100)]
Y = [f(x) for x in range(-100, 100)]
plt.figure(figsize=(20, 10))
plt.plot(X, Y)
plt.show()
```

```python
Y2 = [x ** 2 for x in range(-100, 100)]
plt.figure(figsize=(20, 10))
plt.plot(X, Y2)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.plot(X, Y)
plt.plot(X, Y2)
plt.show()
```
