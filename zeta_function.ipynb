```python
import matplotlib.pyplot as plt
import numpy as np
import mpmath as mp
```

# 課堂範例：將 Critical Line R(Z) = 0.5 映到平面

```python
X = []
Y = []
T = np.linspace(-50, 50, 2000)
for i in T:
    p = mp.zeta(0.5 + i * 1j)
    X.append(p.real)
    Y.append(p.imag)
plt.plot(X, Y)
plt.plot([0.5, 0.5], [min(Y)-0.5, max(Y)+0.5]) # 臨界線
plt.xlim(-2, 4)
plt.ylim(-3, 3)
plt.scatter([0], [0], s=100, c='r')
plt.show()
```

# 作業：將 r = 2 的圓透過 Zeta Function 映到平面

```python
r = 2
a, b = (0, 0)
theta = np.linspace(0, 2 * np.pi, 1000)
X = a + r * np.cos(theta)
Y = b + r * np.sin(theta)
plt.plot(X, Y)
plt.axis('equal')
plt.show()
```

```python
X1 = []
Y1 = []
for x, y in zip(X, Y):
    try:
        z = mp.zeta(x, y)
        X1.append(z.real)
        Y1.append(z.imag)
    except:
        pass
```

```python
plt.plot(X1[:-1], Y1[:-1])    
plt.scatter([0], [0], s=100, c='r')
plt.axis('equal')
plt.show()
```

###### 將圓的座標 (x, y) 代入 Zeta Function ，最後可以得到近似於三角形的圖案。
